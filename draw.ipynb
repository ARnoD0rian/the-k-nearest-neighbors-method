```python
import matplotlib.pyplot as plt
        
def open_file(filename: str) -> list:
    with open(f"results/{filename}.txt", "r") as file:
        data = [int(line) for line in file]
        return data
        
    
```

```python
indexs = [i for i in range(4, 26)]
result_classic = open_file("classic")
result_ClasSwap = open_file("ClasSwap")
result_PopMod = open_file("PopMod")
result_PopModSwap = open_file("PopModSwap")
result_Swap = open_file("Swap")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))
ax.plot(indexs, result_classic, color="blue", linewidth=1.0, linestyle="-")
ax.plot(indexs, result_ClasSwap, color="green", linewidth=1.0, linestyle="-")
ax.plot(indexs, result_PopMod, color="orange", linewidth=1.0, linestyle="-")
ax.plot(indexs, result_PopModSwap, color="black", linewidth=1.0, linestyle="-")
ax.plot(indexs, result_Swap, color="red", linewidth=1.0, linestyle="-")
ax.grid()
ax.set_xlabel(r' ', fontsize=12)
ax.set_ylabel(r'', fontsize=12)
ax.set_title("  ")
ax.legend(["Classic", "ClasSwap", "PopMod (200)", "PopModSwap (50)", "Swap"])
plt.savefig("grafics/end.png")
plt.show()
```
